```python
import pandas as pd
import seaborn as sns
import sqlite3
```

```python
df = pd.read_csv('cell-count.csv')
```

```python
conn = sqlite3.connect("cell_counts.db")
query = "SELECT * FROM samples"
samples = pd.read_sql_query(query, conn)
```

```python
df = samples[["sample", "b_cell", "cd8_t_cell", "cd4_t_cell", "nk_cell", "monocyte"]].set_index("sample")
```

```python
df = df.melt(var_name="population", value_name="count", ignore_index=False)
```

```python
df["total_count"] = df.groupby("sample")["count"].sum()
```

```python
df["percentage"] = df["count"] * 100 / df["total_count"]
```

```python
df = df[["total_count", "population", "count", "percentage"]].sort_values(["sample", "population"])
```
